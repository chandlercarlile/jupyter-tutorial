```python
import pandas as pd

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
headers = ["sepal-length", "sepal-width", "petal-length", "petal-width", "type"]
df = pd.read_csv(url, names = headers)
display(df)
```

```python
from matplotlib import pyplot

df.hist()
pyplot.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df)
```

```python
from sklearn import linear_model

log_model = linear_model.LogisticRegression(max_iter = 1000)

X = df.values[:, 0:4]
y = df.values[:, 4]

log_model.fit(X, y)

print(log_model.predict([[.1, .2, .3, .4]]))
```

```python
from sklearn import metrics

y_pred = log_model.predict(X)
print(metrics.accuracy_score(y, y_pred))
```

```python
from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.3)

log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)

print(metrics.accuracy_score(y_test, y_pred_log))
```

```python
from sklearn import svm

svm_model = svm.SVC(max_iter = 1000)

svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print(metrics.accuracy_score(y_test, y_pred_svm))
```

```python
metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred_log)
```

```python
metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred_svm)
```
